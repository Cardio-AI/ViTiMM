```python
from pathlib import Path
import pandas as pd
```

```python
task = 'phenotyping'
src = Path('results')
```

```python
df_ViTiMM = pd.read_csv(src / f'results_ViTiMM_{task}.csv', index_col=0)
# df_ViTiMM.rename(columns={'stay_id': 'hadm_id'}, inplace=True)
# df_ViTiMM.set_index('hadm_id', inplace=True)
df_ViTiMM.rename(columns={
    'swin_lab_text': 'C',
    'swin_cxr_text': 'X',
    'swin_med_text': 'M',
    'swin_ecg_text': 'E',
    'swin_lab_cxr_text': 'C|X',
    'swin_lab_med_cxr_ecg_text_diagnoses': 'C|M|X|E'
}, inplace=True)
df_ViTiMM.rename(columns={c: f'ViTiMM {c}' for c in df_ViTiMM.columns[2:]}, inplace=True)
df_ViTiMM
```

```python
df_MeTra = pd.read_csv(src / f'results_MeTra_{task}.csv')
df_MeTra.drop(columns=['GT'], inplace=True)
# df_MeTra.set_index('hadm_id', inplace=True)
# metra_meta = pd.read_csv('results/MeTra_test_metadata.csv', index_col=0)
# metra_meta = metra_meta[['stay_id','y_true']]
# metra_meta.rename(columns={'stay_id': 'hadm_id'}, inplace=True)
# df_MeTra = df_MeTra.merge(metra_meta, on='hadm_id', how='left')
# df_MeTra.set_index('hadm_id', inplace=True)
df_MeTra.rename(columns={
    'EHR': 'C',
    'CXR': 'X',
    'EHR_CXR': 'C|X'
}, inplace=True)
df_MeTra.rename(columns={c: f'MeTra {c}' for c in df_MeTra.columns[2:]}, inplace=True)
df_MeTra
```

```python
df_MedFuse = pd.read_csv(src / f'results_MedFuse_{task}.csv')
# df_MedFuse.set_index('hadm_id', inplace=True)
df_MedFuse.drop(columns=['GT'], inplace=True)
df_MedFuse.rename(columns={
    'uni_ehr_paired': 'C',
    'uni_cxr_paired': 'X',
    'medFuse_paired': 'C|X'
}, inplace=True)
df_MedFuse.rename(columns={c: f'MedFuse {c}' for c in df_MedFuse.columns[2:]}, inplace=True)
# df_MedFuse.rename(columns={'MedFuse GT': 'y_true'}, inplace=True)
df_MedFuse
```

```python
df_MeTra = df_MeTra.reset_index()
df_MedFuse = df_MedFuse.reset_index()
df_ViTiMM = df_ViTiMM.reset_index()
df = df_ViTiMM.merge(df_MedFuse, on=['hadm_id','class_name'], how='inner')
df = df.merge(df_MeTra, on=['hadm_id','class_name'], how='inner')
df
```

```python
from scipy.stats import ttest_rel
# dfp = df.drop(columns=['y_true']).set_index('hadm_id')

cols = [
    'MeTra C', 
    'MeTra X', 
    'MeTra C|X',
    'MedFuse C', 
    'MedFuse X',
    'MedFuse C|X', 
    'ViTiMM C', 
    'ViTiMM X', 
    'ViTiMM M', 
    'ViTiMM E', 
    'ViTiMM C|X',
    'ViTiMM C|M|X|E'
]
dfp = df[cols]
methods = dfp.columns
p_values = pd.DataFrame(index=methods, columns=methods)

for method_1 in methods:
    for method_2 in methods:
        if method_1 != method_2:
            _df = df.dropna(subset=[method_1,method_2])
            _, p_val = ttest_rel(_df[method_1], _df[method_2])
            p_values.loc[method_1, method_2] = p_val
        else:
            p_values.loc[method_1, method_2] = None  # No test against itself
p_values = p_values.astype(float)
p_values.round(3)#.to_latex()
```
